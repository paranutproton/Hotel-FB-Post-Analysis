```python
new_column_names = []
for col_name in df.columns:
  new_col_name = ''
  for char in col_name:
    if (char.isspace()):
        char = '_'
    elif (not char.isalnum()):
        continue
    new_col_name = new_col_name + char.lower()
  new_column_names.append(new_col_name)
```

```python
df_no_paid['str_date'] = df_no_paid['post_date'].dt.strftime('%a %y-%m-%d')
df_no_paid.head()
```

```python
df_no_paid['dow'] = df_no_paid['post_date'].dt.dayofweek.astype(str) + '_' + df_no_paid['str_date'].str[:3]
df_no_paid.head()
```

```python
df_no_paid.groupby(['type','dow'])['lifetime_post_organic_reach'].sum().unstack().fillna(0).astype(int)
```

```python
df_no_paid['hour'] = df_no_paid['post_date'].dt.hour
mean_likes_by_dow_time = pd.pivot_table(df_no_paid, index='hour', columns='dow',values='lifetime_post_organic_reach',aggfunc='mean', fill_value=0).astype(int)
mean_likes_by_dow_time
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(8,6))
ax = sns.heatmap(mean_likes_by_dow_time, ax=ax, annot=True, fmt='d')
ax.set_title('Mean lifetime_post_organic_reach by Date of Week and Time of Day')
plt.show()
### TIME is in California Pacific Time Zone
```
